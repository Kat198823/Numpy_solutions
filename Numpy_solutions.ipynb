```python
#q. Получить сумму всех значений в стартовой матрице mat
import numpy as np
mat = np.arange(1,26).reshape(5,5)
print(mat,"\n")
a = mat.sum()
print(a,"\n")
```

```python
#r. Получить стандартное отклонение в стартовой матрице mat
import numpy as np
mat = np.arange(1,26).reshape(5,5).std()
print(mat,"\n")
```

```python
#s. Получить сумму значений в колонках в стартовой матрице mat
import numpy as np
mat = np.arange(1,26).reshape(5,5)
print(mat,"\n")
a = mat.sum(axis = 0)
print(a,"\n")
```
